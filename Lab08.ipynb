```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
t, v = np.loadtxt('droptower_vdata.txt', unpack = True)
x = integrate.cumtrapz(v, t, initial = 30) 
a = np.diff(v)/np.diff(t)
#below makes it so that at time 0 the acceleration is 0 since the ride starts at rest
a = np.append([0], a)
```

```python
plt.plot(t,v, color = 'r', label = 'velocity')
plt.plot(t, x, color = 'b', label = 'position')
plt.plot(t, a, color = 'g', label = 'acceleration')
plt.legend()
```

```python
a_pos = a>0
pos_accel = a[a_pos]
average_accel = np.average(pos_accel)
average_accel
```
